```python
# Student Management System using OOP

class Student:
    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks  # dictionary of subject: marks

    def calculate_grade(self):
        avg = sum(self.marks.values()) / len(self.marks)
        if avg >= 90:
            return "A+"
        elif avg >= 75:
            return "A"
        elif avg >= 60:
            return "B"
        elif avg >= 40:
            return "C"
        else:
            return "Fail"

    def display_info(self):
        print(f"\nRoll No: {self.roll_no}")
        print(f"Name: {self.name}")
        print("Marks:")
        for subject, mark in self.marks.items():
            print(f"  {subject}: {mark}")
        print(f"Grade: {self.calculate_grade()}")


class StudentManagementSystem:
    def __init__(self):
        self.students = []

    def add_student(self, student):
        self.students.append(student)
        print(f"\n✅ Student {student.name} added successfully!")

    def view_students(self):
        if not self.students:
            print("\nNo students found.")
        else:
            print("\n=== Student Records ===")
            for student in self.students:
                student.display_info()

    def search_student(self, roll_no):
        for student in self.students:
            if student.roll_no == roll_no:
                print("\n=== Student Found ===")
                student.display_info()
                return
        print("\n❌ Student not found.")


# Main Program
def main():
    sms = StudentManagementSystem()

    while True:
        print("\n===== Student Management System =====")
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student by Roll No")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            roll_no = input("Enter Roll No: ")
            name = input("Enter Name: ")
            marks = {}
            subjects = ["Math", "Science", "English"]
            for subject in subjects:
                marks[subject] = int(input(f"Enter marks for {subject}: "))
            student = Student(roll_no, name, marks)
            sms.add_student(student)

        elif choice == "2":
            sms.view_students()

        elif choice == "3":
            roll_no = input("Enter Roll No to search: ")
            sms.search_student(roll_no)

        elif choice == "4":
            print("\nExiting... Thank you!")
            break

        else:
            print("\nInvalid choice. Try again!")

if __name__ == "__main__":
    main()
```
